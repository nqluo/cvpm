```python

from scipy.io import loadmat
import matplotlib.pyplot as plt

rppg_data_path = 'rppg/2023-06-11-18-54.mat' # .mat file path
rppg_data = loadmat(rppg_data_path)
rppg_data.keys()
```

```python
time = rppg_data['time'].reshape(-1)
rppg = rppg_data['rPPG'].reshape(-1)
plt.figure(figsize=(8, 5))
plt.plot(time, rppg)
plt.xlabel('Time (s)')
plt.ylabel('rPPG Amplitude (a.u.)')
plt.title('rPPG')
```
