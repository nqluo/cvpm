```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
rppg_data_path = 'demo/demo1.mat' # .mat file path
rppg_data = loadmat(rppg_data_path)
rppg_data.keys()
```

```python
time = rppg_data['time'].reshape(-1)
r = rppg_data['r'].reshape(-1)
g = rppg_data['g'].reshape(-1)
b = rppg_data['b'].reshape(-1)
rppg = rppg_data['rPPG'].reshape(-1)
```

```python
# plot RGB signal   
plt.figure(figsize=(8, 5))
plt.plot(time, r, color = 'red', label='r')
plt.plot(time, g, color = 'green', label='g')
plt.plot(time, b, color = 'blue', label='b')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (a.u.)')
plt.title('RGB')
plt.legend();
```

```python
#plot rPPG signal
plt.figure(figsize=(8, 5))
plt.plot(time, rppg, label='rPPG')
plt.xlabel('Time (s)')
plt.ylabel('rPPG Amplitude (a.u.)')
plt.title('rPPG')
plt.legend();
```
